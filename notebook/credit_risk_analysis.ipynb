```python
def hist_box_plot(data,title='',xlabel='',ylabel=''):
    plt.figure(figsize=(9,4))
    plt.subplot(1,2,1)
    plt.title(title)
    plt.hist((data))
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.subplot(1,2,2)
    plt.title(title)
    sns.boxplot(data=(data))
    plt.show()

def bar_plot(data,title='',xlabel='',ylabel=''):
    counts = data.value_counts()
    plt.title(title)
    plt.bar(counts.index,counts.values)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
bins = [0,25,35,50,100]
labels = ['Youth','Young Adult','Adult','Elderly']
df['Age_bin'] = pd.cut(df['Age'],bins = bins,labels=labels)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['Risk','Age'],axis=1)
y = df['Risk']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=23)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,LabelEncoder

ordinal_columns = ['Housing','Saving accounts','Checking account']
nominal_columns = ['Sex','Job','Purpose','Age_bin']

ordinal_encoder = OrdinalEncoder()
one_hot_encoder = OneHotEncoder(sparse_output=False,handle_unknown='ignore',drop='first')
label_encoder = LabelEncoder()

X_train[ordinal_columns] = ordinal_encoder.fit_transform(X_train[ordinal_columns])
X_test[ordinal_columns] = ordinal_encoder.transform(X_test[ordinal_columns])


encoded_nominal_train = one_hot_encoder.fit_transform(X_train[nominal_columns])
encoded_nominal_train_df = pd.DataFrame(encoded_nominal_train,columns=one_hot_encoder.get_feature_names_out(nominal_columns))

X_train.drop(columns=nominal_columns,inplace=True)
X_train = pd.concat([X_train.reset_index(drop=True),encoded_nominal_train_df.reset_index(drop=True)],axis=1)

encoded_nominal_test = one_hot_encoder.transform(X_test[nominal_columns])
encoded_nominal_test_df = pd.DataFrame(encoded_nominal_test,columns=one_hot_encoder.get_feature_names_out(nominal_columns))

X_test.drop(columns=nominal_columns,inplace=True)
X_test = pd.concat([X_test.reset_index(drop=True),encoded_nominal_test_df.reset_index(drop=True)],axis=1)

y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)

Xy_train = X_train.copy()
Xy_train['Risk'] = y_train
Xy_test = X_test.copy()
Xy_test['Risk'] = y_test
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(Xy_train.corr(),annot=True,cmap='Blues',fmt='.2f',linewidths=0.5)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report,confusion_matrix,f1_score
models = {
    'logisticModel' : LogisticRegression(max_iter=1000),
    'rfModel' : RandomForestClassifier(random_state=42),
    'xgbModel' : XGBClassifier(use_label_encoder=False,eval_metric = 'logloss',random_state=42)
}

for name,model in models.items():
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test,y_pred)
    sns.heatmap(cm,annot=True,fmt='.2f')
    plt.show()
    print(f'Model : {name}')
    print(f'F1-Score: {f1_score(y_test,y_pred)}')
    print(classification_report(y_test,y_pred))
```
