```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("..\\data\\german_credit_data.csv")
df.head()
```

```python
df.drop('Unnamed: 0',axis =1,inplace=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.title("Age")
plt.hist((df['Age']))
plt.xlabel("Age")
plt.ylabel("Count")

plt.subplot(1,2,2)
plt.title("Age")
sns.boxplot(data=(df['Age']))
plt.show()
```

```python
df['Sex'].value_counts().
```

```python
sex_counts = df['Sex'].value_counts()
plt.title("Sex")
plt.bar(sex_counts.index,sex_counts.values)
plt.xlabel("Distribusi Sex")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.title("Job")
plt.hist((df['Job']))
plt.xlabel("Job")
plt.ylabel("Count")
plt.show()
```

```python
Housing_counts = df['Housing'].value_counts()
plt.title("Housing")
plt.bar(Housing_counts.index,Housing_counts.values)
plt.xlabel("Distribusi Housing")
plt.ylabel("Count")
plt.show()
```

```python
Savingaccounts_counts = df['Saving accounts'].value_counts()
plt.title("Saving accounts")
plt.bar(Savingaccounts_counts.index,Savingaccounts_counts.values)
plt.xlabel("Distribusi Saving accounts")
plt.ylabel("Count")
plt.show()
```

```python
Checkingaccount_counts = df['Checking account'].value_counts()
plt.title("Checking account")
plt.bar(Checkingaccount_counts.index,Checkingaccount_counts.values)
plt.xlabel("Distribusi Checking account")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.title("Credit amount")
plt.hist((df['Credit amount']))
plt.xlabel("Credit amount")
plt.ylabel("Count")

plt.subplot(1,2,2)
plt.title("Credit amount")
sns.boxplot(data=(df['Credit amount']))
plt.show()
```

```python
df['Checking account'].mode()[0]
```

```python
plt.figure(figsize=(12,6))
df['Saving accounts'].fillna('little',inplace=True)
df['Checking account'].fillna(df['Checking account'].mode()[0],inplace=True)
plt.subplot(1,2,1)
plt.bar(Savingaccounts_counts.index,Savingaccounts_counts.values)
plt.title('Saving accounts')

plt.subplot(1,2,2)
plt.bar(Checkingaccount_counts.index,Checkingaccount_counts.values)
plt.title('Checking accounts')
plt.show()
```
